## 2. Transformer
In the first chapter, the attention-Seq2Seq model is implemented, which has the advantages that it elminates the bottleneck of fixed length encoding, and all the information of the encoding can be transmitted to decoder without any loss. While the normal Seq2Seq model compress all the encoding information to a sigle vector, then this vector will be the only input tot the decoder. Nevertheless, Attention-Seq2Seq model also has some drawbacks, because it utilizes GRU/LSTM, it leads to a low parallelism; besides, attention only exists between encoder and decoder while there is not self-attention.

Based on the above analysis, in this chapter, transformer model is imlemented, which reserves self-attetnion layer and removes the RNN structure. In other words, it is a model of attention without RNN. Besides, it also iptimizes the normal attention model in some asprects. The original strcuture of transformer is shown is the following figure.

The features in transformer model is listed as follows:
* Encoder-Decoder structure
* Multiple Encoder-Decoder layers
* positional encoding
* multi-head attention
* Add (Resnet) & norm (normalization) layer
